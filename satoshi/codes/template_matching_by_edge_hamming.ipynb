```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
img = cv2.imread('../images/music1.png',0)
plt.imshow(img,cmap = 'gray')
```

```python
img_edge = cv2.Canny(img,100,200)#0 0r 255
plt.imshow(img_edge,cmap = 'gray')
```

```python
template2 = cv2.imread('../images/template2.png',0)
plt.imshow(template2,cmap = 'gray')
```

```python
template2_edge = cv2.Canny(template2,100,200)#0 0r 255
plt.imshow(template2_edge,cmap = 'gray')
```

```python
img_=img/255
template=template2

m=len(template)
n=len(template[0])

M=len(img)
N=len(img[0])

hamming_sim=np.zeros_like(img)

for i in xrange(M):
    print i,
    for j in xrange(N):      
        for k in xrange(m):
            for l in xrange(n):
                if i+k in xrange(M) and j+l in xrange(N):
                    hamming_sim[i][j]+=img_[(i+k,j+l)]*template[(k,l)]+(1-img_[(i+k,j+l)])*(1-template[(k,l)])
```

```python
np.save('../data/hamming_sim.np',hamming_sim)
```

```python
plt.imshow(hamming_sim)
cv2.imwrite('../images/music1_template1.png',hamming_sim)
plt.imshow(hamming_sim, cmap=plt.cm.hot)
plt.colorbar()
```
